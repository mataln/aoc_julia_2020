```julia
function is_valid_1(password::AbstractString)
    """
    Takes a single password & policy
    string and checks whether it is 
    valid
    """
    chunks = split(password, r"\s|-")
    lower, upper, letter, passkey = parse(Int64, chunks[1]), parse(Int64, chunks[2]), chunks[3][1], chunks[4]
    
    count = 0
    for chr in passkey
        chr == letter && (count+=1)
    end
    
    return lower <= count <= upper
end

function is_valid_2(password::AbstractString)
    """
    Takes a single password & policy
    string and checks whether it is 
    valid
    """
    chunks = split(password, r"\s|-")
    left, right, letter, passkey = parse(Int64, chunks[1]), parse(Int64, chunks[2]), chunks[3][1], chunks[4]
    
    return (passkey[left] == letter)  (passkey[right] == letter)
end

function count_valid(passwords::String; task=1)
    """
    Takes the entire list of passwords + policies 
    (separated by line break) as a single string 
    and returns the number of valid passwords
    """
    passwords_list = split(passwords, "\n")
    valid_passwords = 0
    
    for password in passwords_list
        if task==1
            is_valid_1(password) && (valid_passwords += 1)
        else
            is_valid_2(password) && (valid_passwords += 1)
        end
    end
        
    return valid_passwords
end
```

```julia
using Test

@testset "Task 1" begin
    @test is_valid_1("1-3 a: abcde") == true
    @test is_valid_1("1-3 b: cdefg") == false
    @test is_valid_1("2-9 c: ccccccccc") == true
    @test count_valid(sample_input) == 2
    @test count_valid(test_input) == 500
end
    
@testset "Task 2" begin
    @test is_valid_2("1-3 a: abcde") == true
    @test is_valid_2("1-3 b: cdefg") == false
    @test is_valid_2("2-9 c: ccccccccc") == false
    @test count_valid(sample_input, task = 2) == 1
    @test count_valid(test_input, task = 2) == 313
end
        
```
