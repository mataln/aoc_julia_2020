```julia
task2_gradients = [(1,1),(3,1),(5,1),(7,1),(1,2)]
```

```julia
function task1(puzzle_input::String, right::Int64, down::Int64)
    """
    Counts number of encountered trees (#) in puzzle
    input when traversing the gradient given in steps 
    by right, down
    """
    down <=0 && throw(DomainError(down, "Fuck off"))
    
    puzzle_matrix = split(puzzle_input) #It's not actually a matrix, it's a list of strings. Close enough.
    height, width = length(puzzle_matrix), length(puzzle_matrix[1])
    
    x, j, i, tree_count = 1, 1, 1, 0
    while i <= height
        puzzle_matrix[i][j] == '#' && (tree_count += 1)
        i += down
        x += right
        j = (x+width-1)%width + 1      
    end
    return tree_count
end

function task2(puzzle_input::String, gradients::Array{Tuple{Int64,Int64},1})
    """
    Performs task 2 for the list of gradients (as 
    (right, down)) tuples given in gradient_list 
    """
    prod = 1
    for gradient in gradients
        prod *= task1(puzzle_input, gradient[1], gradient[2])
    end
    return prod
end
```

```julia
using Test

@testset "Task 1" begin
    @test task1(sample_input, 0, 1) == 3
    @test task1(sample_input, 3, 1) == 7
    @test task1(test_input, 3, 1) == 169
    @test_throws DomainError task1(sample_input, 1, 0)
end

@testset "Task 2" begin
    @test task2(sample_input, task2_gradients) == 336
    @test task2(test_input, task2_gradients) == 7560370818
end
```
