```julia
using Base.Cartesian

macro task_n(puzzle_input::Array{Int64,1}, n=2::Int64)
    exp = quote
        @nloops $n i _ -> 1:length($puzzle_input) begin           
            sum = 0               
            for j in @ntuple $n i 
                sum += $puzzle_input[j]
            end               
            if sum == 2020
                prod = 1
                for j in @ntuple $n i 
                    prod *= $puzzle_input[j]
                end 
                return prod
            end  
        end
    end
    eval(exp)
end
```

```julia
using Test

@testset "Example input - n=2" begin
    @test task_n(sample_input, 2) == 514579 #Check it works on the example input
end

@testset "Example input - n=3" begin
    @test task_n(sample_input, 3) == 241861950 #Check it works on the example input
end
```
